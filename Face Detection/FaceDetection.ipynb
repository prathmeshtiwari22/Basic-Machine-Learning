```python
import cv2
from google.colab.patches import cv2_imshow  # Import the Colab patch

def face_detect(img_dir):
    img = cv2.imread(img_dir)

    # Check if image loaded successfully
    if img is None:
        print("Error: Could not load image. Check the file path and permissions.")
        return None

    img_g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    FaceRecog = cv2.CascadeClassifier('/content/drive/MyDrive/Machine Learning Projects/Face Detection/face_recog_data.xml')
    cor = FaceRecog.detectMultiScale(img_g,1.3,5)

    try:
        if cor.all():
            for (x,y,w,h) in cor:
                img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 3)
            return img
    except AttributeError:
        return img

# Verify the path to your image
img_path = '/content/drive/MyDrive/Machine Learning Projects/Face Detection/p1.jpg'  # Make sure this is the correct path
result_img = face_detect(img_path)

if result_img is not None:
    cv2_imshow(result_img)  # Use cv2_imshow instead of cv2.imshow
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
